Task 8: Investment Duration

```python
def convert_duration_to_years(Duration):
    if Duration == "less than 1 years":
        return 0.5
    elif Duration == "1-3 years":
        return 2
    elif Duration == "more than5  years":
        return 4
    elif Duration == "more than 5 years":
        return 6
    else:
        return  None
```

```python
investment_duration =df['Duration']
```

```python
numerical_durations = investment_duration.apply(convert_duration_to_years)
numerical_durations
```

```python
average_duration = numerical_durations.mean()
average_duration
```

```python
print(f"average investment duration: {average_duration:.2f}years")
```
